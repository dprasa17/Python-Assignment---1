Question-3.Compare and contrast mutable and immutable objects in Python with examples.

Answer-3. In Python, objects can be classified into two categories: mutable and immutable.

Immutable Objects

Immutable objects are those that cannot be modified after creation. Examples include:

- Integers (int)
- Floating-point numbers (float)
- Strings (str)
- Tuples (tuple)

Here's an example with strings:


my_string = "Hello"
try:
    my_string[0] = "h"  # Trying to modify the string
except TypeError as e:
    print(e)  # Output: 'str' object does not support item assignment


As you can see, attempting to modify a string results in a TypeError.

Mutable Objects

Mutable objects, on the other hand, can be modified after creation. Examples include:

- Lists (list)
- Dictionaries (dict)
- Sets (set)

Here's an example with lists:


my_list = [1, 2, 3]
my_list[0] = 10  # Modifying the list
print(my_list)  # Output: [10, 2, 3]


In this case, modifying the list is allowed.

Key differences

Here are the key differences between mutable and immutable objects:

- Modifiability: Immutable objects cannot be modified, while mutable objects can be modified.
- Memory allocation: When you modify an immutable object, a new object is created in memory, whereas modifying a mutable object changes the existing object in memory.
- Hashability: Immutable objects can be used as dictionary keys, while mutable objects cannot (because their hash value would change after modification).

Here's an example that demonstrates the difference in memory allocation:


# Immutable object (string)
s1 = "Hello"
s2 = s1
s1 += " World"
print(id(s1))  # Different memory address
print(id(s2))  # Same memory address as original string

# Mutable object (list)
l1 = [1, 2, 3]
l2 = l1
l1.append(4)
print(id(l1))  # Same memory address
print(id(l2))  # Same memory address as l1


In the case of the immutable string, modifying s1 creates a new object in memory. For the mutable list, modifying l1 changes the existing object in memory, which is also referenced by l2.

Question-4.Discuss the different types of operators in Python and provide examples of how they are used.

Answer-4.Python has several types of operators that can be used to perform various operations. Here are the main types of operators in Python:

1. Arithmetic Operators
Arithmetic operators are used to perform mathematical operations.

- + (addition)
- - (subtraction)
- * (multiplication)
- / (division)
- // (floor division)
- % (modulus)
- ** (exponentiation)

Examples:


a = 10
b = 3

print(a + b)  # Output: 13
print(a - b)  # Output: 7
print(a * b)  # Output: 30
print(a / b)  # Output: 3.3333333333333335
print(a // b)  # Output: 3
print(a % b)  # Output: 1
print(a ** b)  # Output: 1000


2. Comparison Operators
Comparison operators are used to compare values.

- == (equal to)
- != (not equal to)
- > (greater than)
- < (less than)
- >= (greater than or equal to)
- <= (less than or equal to)

Examples:


a = 10
b = 5

print(a == b)  # Output: False
print(a != b)  # Output: True
print(a > b)  # Output: True
print(a < b)  # Output: False
print(a >= b)  # Output: True
print(a <= b)  # Output: False


3. Logical Operators
Logical operators are used to perform logical operations.

- and (logical and)
- or (logical or)
- not (logical not)

Examples:


a = True
b = False

print(a and b)  # Output: False
print(a or b)  # Output: True
print(not a)  # Output: False


4. Assignment Operators
Assignment operators are used to assign values to variables.

- = (assignment)
- += (addition assignment)
- -= (subtraction assignment)
- *= (multiplication assignment)
- /= (division assignment)
- //= (floor division assignment)
- %= (modulus assignment)
- **= (exponentiation assignment)

Examples:


a = 10

a += 5  # equivalent to a = a + 5
print(a)  # Output: 15

a -= 3  # equivalent to a = a - 3
print(a)  # Output: 12

a *= 2  # equivalent to a = a * 2
print(a)  # Output: 24


5. Bitwise Operators
Bitwise operators are used to perform bitwise operations on integers.

- & (bitwise and)
- | (bitwise or)
- ^ (bitwise xor)
- ~ (bitwise not)
- << (left shift)
- >> (right shift)

Examples:


a = 5  # binary: 101
b = 3  # binary: 011

print(a & b)  # Output: 1 (binary: 001)
print(a | b)  # Output: 7 (binary: 111)
print(a ^ b)  # Output: 6 (binary: 110)
print(~a)  # Output: -6 (binary: ...111010)
print(a << 1)  # Output: 10 (binary: 1010)
print(a >> 1)  # Output: 2 (binary: 10)


6. Membership Operators
Membership operators are used to check if a value is present in a sequence.

- in (membership test)
- not in (non-membership test)

Examples:


fruits = ['apple', 'banana', 'cherry']

print('apple' in fruits)  # Output: True
print('grape' in fruits)  # Output: False
print('apple' not in fruits)  # Output: False


7. Identity Operators
Identity operators are used to check if two variables refer to the same object.

- is (identity test)
- is not (non-identity test)

Examples:


a = [1, 2, 3]
b = [1, 2, 3]
c = a

print(a is b)  # Output: False
print(a is c)  # Output: True
print(a == b)  # Output: True


These are the different types of operators in python.

Question-5.Explain the concept of type casting in Python with examples.

Answer-5.Type casting in Python is the process of converting one data type into another. Python supports two types of type casting:

1. Implicit Type Casting
Implicit type casting is automatically performed by Python when it encounters a mix of data types in an expression. Python converts the data type of one operand to match the data type of the other operand.

Examples:


num_int = 10
num_float = 5.5

result = num_int + num_float
print(result)  # Output: 15.5
print(type(result))  # Output: <class 'float'>

num_int = 10
num_complex = 3 + 4j

result = num_int + num_complex
print(result)  # Output: (13+4j)
print(type(result))  # Output: <class 'complex'>


In these examples, Python automatically converts the integer to a float or complex number to perform the operation.

2. Explicit Type Casting
Explicit type casting is performed manually by the programmer using built-in functions like int(), float(), str(), complex(), bool(), etc.

Examples:


# Convert string to integer
num_str = "10"
num_int = int(num_str)
print(num_int)  # Output: 10
print(type(num_int))  # Output: <class 'int'>

# Convert integer to float
num_int = 10
num_float = float(num_int)
print(num_float)  # Output: 10.0
print(type(num_float))  # Output: <class 'float'>

# Convert float to string
num_float = 10.5
num_str = str(num_float)
print(num_str)  # Output: 10.5
print(type(num_str))  # Output: <class 'str'>

# Convert integer to complex
num_int = 10
num_complex = complex(num_int)
print(num_complex)  # Output: (10+0j)
print(type(num_complex))  # Output: <class 'complex'>

# Convert string to boolean
bool_str = "True"
bool_val = bool(bool_str)
print(bool_val)  # Output: True
print(type(bool_val))  # Output: <class 'bool'>


In these examples, we use built-in functions to explicitly convert one data type to another.

Common Type Casting Functions:

- int(): Converts a value to an integer.
- float(): Converts a value to a floating-point number.
- str(): Converts a value to a string.
- complex(): Converts a value to a complex number.
- bool(): Converts a value to a boolean.
- list(), tuple(), set(), dict(): Convert a value to a list, tuple, set, or dictionary, respectively.

Important Notes:

- When converting a string to a number (int or float), the string must be a valid numeric literal.
- When converting a float to an integer, the decimal part is truncated.
- When converting a value to a boolean, the following values are considered False: 0, 0.0, "", [], (), {}, None. All other values are considered True.

Question-6.How do conditional statements work in Python? Illustrate with example.

Answer-6.Conditional statements in Python are used to execute different blocks of code based on specific conditions. These conditions are evaluated as True or False, and the corresponding block of code is executed.

Types of Conditional Statements
1. If Statement: Used to execute a block of code if a condition is True.
2. If-Else Statement: Used to execute one block of code if a condition is True and another block if the condition is False.
3. If-Elif-Else Statement: Used to check multiple conditions and execute different blocks of code based on the conditions.
4. Nested If Statement: Used to check conditions within conditions.

Examples
1. If Statement

x = 10
if x > 5:
    print("x is greater than 5")


In this example, the condition x > 5 is evaluated as True, so the code inside the if block is executed.

2. If-Else Statement

x = 10
if x > 15:
    print("x is greater than 15")
else:
    print("x is less than or equal to 15")


In this example, the condition x > 15 is evaluated as False, so the code inside the else block is executed.

3. If-Elif-Else Statement

x = 10
if x > 15:
    print("x is greater than 15")
elif x == 10:
    print("x is equal to 10")
else:
    print("x is less than 10")


In this example, the condition x > 15 is evaluated as False, but the condition x == 10 is evaluated as True, so the code inside the elif block is executed.

4. Nested If Statement

x = 10
y = 5
if x > 5:
    if y > 2:
        print("Both conditions are true")
    else:
        print("x is greater than 5, but y is not greater than 2")
else:
    print("x is not greater than 5")


In this example, both conditions x > 5 and y > 2 are evaluated as True, so the code inside the nested if block is executed.

Conditional Expressions
Python also supports conditional expressions, which are a concise way to write simple if-else statements.


x = 10
result = "x is greater than 5" if x > 5 else "x is less than or equal to 5"
print(result)


In this example, the condition x > 5 is evaluated as True, so the value "x is greater than 5" is assigned to the variable result.

Question-7.Describe the different types of loops in Python and their use cases with examples.

Answer-7.Python has two primary types of loops: For Loops and While Loops.

1. For Loops
-For loops are used to iterate over a sequence (such as a list, tuple, dictionary, set, or string) and execute a block of code for each item in the sequence.

Example:


fruits = ['apple', 'banana', 'cherry']
for fruit in fruits:
    print(fruit)


In this example, the loop iterates over the fruits list and prints each fruit.

Use Cases:

- Iterating over a list or tuple to perform operations on each item.
- Looping through a dictionary to access key-value pairs.
- Iterating over a string to process each character.

2. While Loops

-While loops are used to execute a block of code as long as a certain condition is true.

Example:


i = 0
while i < 5:
    print(i)
    i += 1


In this example, the loop continues to execute until i is no longer less than 5.

Use Cases:

- Repeating a task until a certain condition is met.
- Implementing a game loop or simulation.
- Reading input from a user until a specific input is received.

Loop Control Statements

-Python provides several loop control statements to manage the flow of loops:

- break: Exits the loop immediately.
- continue: Skips the current iteration and moves to the next one.
- pass: Does nothing and moves to the next statement.

Example:


fruits = ['apple', 'banana', 'cherry']
for fruit in fruits:
    if fruit == 'banana':
        continue
    print(fruit)


In this example, the loop skips the 'banana' fruit and prints the other fruits.

Nested Loops

-Python allows nesting loops inside other loops. This is useful for iterating over multiple sequences or performing complex operations.

Example:


colors = ['red', 'green', 'blue']
shapes = ['circle', 'square', 'triangle']
for color in colors:
    for shape in shapes:
        print(f"{color} {shape}")
